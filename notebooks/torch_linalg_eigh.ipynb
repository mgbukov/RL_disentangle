```python
import numpy as np
import torch

from src.stategen import sample_haar_full
from src.quantum_env import rdm_2q_complex
```

```python
np.random.seed(5)
psi = sample_haar_full(3)
psi
```

```python
rdms = rdm_2q_complex(np.expand_dims(psi, 0))
rdms.shape
```

```python
x = psi.reshape(1, 4, -1)
rdms = x @ np.transpose(x.conj(), [0,2,1])
rdms += np.finfo(rdms.dtype).eps * np.diag([0.0, 1.0, 2.0, 4.0])
rdms
```

```python
numpy_rhos, numpy_Us = np.linalg.eigh(rdms)
numpy_Us.round(4)
```

```python
torch_rhos, torch_Us = torch.linalg.eigh(torch.from_numpy(rdms))
torch_Us.numpy().round(4)
```
